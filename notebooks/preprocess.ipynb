#### Load Data

```python
# Load JSON data into a dataframe
import pandas as pd
import json


def load_data():
    with open(r'..\data\scrape_results\dashen\review_dashen_bank_scrape_result.json', 'r', encoding='utf-8') as dashed_data:
        dashen_json = json.load(dashed_data)
        df_dashen = pd.DataFrame(dashen_json)
    with open(r'..\data\scrape_results\boa\review_boa_bank_scrape_result.json', 'r', encoding='utf-8') as boa_data:
        boa_json = json.load(boa_data)
        df_boa = pd.DataFrame(boa_json)
    with open(r'..\data\scrape_results\cbe\review_cbe_bank_scrape_result.json', 'r', encoding='utf-8') as cbe_data:
        cbe_json = json.load(cbe_data)
        df_cbe = pd.DataFrame(cbe_json)
    return df_dashen, df_boa, df_cbe

df_dashen, df_boa, df_cbe = load_data()
    
```

```python
def get_info():
    print("------------------------------------")
    print("CBE BANK INFO")
    df_cbe.info()
    print("------------------------------------")
    print("BOA BANK INFO")
    df_boa.info()
    print("------------------------------------")
    print("DASHEN BANK INFO")
    df_dashen.info()

get_info()
```

```python
def null_percentage(df):
    number_of_rows, number_of_columns = df.shape
    df_size = number_of_rows * number_of_columns
    
    null_size = (df.isnull().sum()).sum()
    percentage = round((null_size / df_size) * 100, 2)
    return f"Data Frame contain total null values of { percentage }%"

def calculate_total_null_values():
    print("------------------------------------")
    print("CBE BANK NULL VALUES")
    result = null_percentage(df_cbe)
    print(f"CBE dataframe: {result}")
    print("------------------------------------")
    print("BOA BANK NULL VALUES")
    result = null_percentage(df_boa)
    print(f"BOA dataframe: {result}")
    print("------------------------------------")
    print("DASHEN BANK NULL VALUES")
    result = null_percentage(df_dashen)
    print(f"DASHEN dataframe: {result}")
calculate_total_null_values()
```

```python
def missing_values_per_column(df):
    return df.isnull().sum()

def calculate_missing_values_per_column():
    print("------------------------------------")
    print("CBE BANK COLUMN MISSING VALUES")
    result = missing_values_per_column(df_cbe)
    print(f"CBE dataframe: {result}")
    print("------------------------------------")
    print("BOA BANK COLUMN MISSING VALUES")
    result = missing_values_per_column(df_boa)
    print(f"BOA dataframe: {result}")
    print("------------------------------------")
    print("DASHEN BANK COLUMN MISSING VALUES")
    result = missing_values_per_column(df_dashen)
    print(f"DASHEN dataframe: {result}")

calculate_missing_values_per_column()
```

As can be seen from the above output, most missing values are concentrated on columns
replyContent
repliedAt
appVersion

Hence we will remove these columns and recalculate the missing values percentage and check
if its less than 5%

#### Dropping columns with high numer of missing values

```python
df_cbe = df_cbe.drop(['replyContent', 'repliedAt', 'appVersion'], axis=1)
df_boa = df_boa.drop(['replyContent', 'repliedAt', 'appVersion'], axis=1)
df_dashen = df_dashen.drop(['replyContent', 'repliedAt', 'appVersion'], axis=1)
```

#### Re-calculating total null values

```python
calculate_total_null_values()
```

#### Check for duplicates

```python
def check_for_duplicates(df):
    duplicates_df = df[df.duplicated()]
    return duplicates_df

def check_duplicates_for_all_banks():
    print("------------------------------------")
    print("CBE BANK DUPLICATES")
    result = check_for_duplicates(df_cbe)
    print(f"CBE duplicates dataframe: {result}")
    print("------------------------------------")
    print("BOA BANK DUPLICATES")
    result = check_for_duplicates(df_boa)
    print(f"BOA duplicates dataframe: {result}")
    print("------------------------------------")
    print("DASHEN BANK DUPLICATES")
    result = check_for_duplicates(df_dashen)
    print(f"DASHEN duplicates dataframe: {result}")

check_duplicates_for_all_banks()
```

```python
def normalize_date(df):
    # Convert 'at' column to datetime object
    df['at'] = pd.to_datetime(df['at'], errors='coerce')
    df['created_at_date'] = df['at'].dt.normalize()

def normalize_dates_for_all_banks():
    normalize_date(df_cbe)
    normalize_date(df_boa)
    normalize_date(df_dashen)

normalize_dates_for_all_banks()
```
