#### Preprocess Dashen Bank Data

```python
# Load JSON data into a dataframe
import pandas as pd
import json


def load_data():
    with open(r'..\data\scrape_results\dashen\review_dashen_bank_scrape_result.json', 'r', encoding='utf-8') as dashed_data:
        dashen_json = json.load(dashed_data)
        df_dashen = pd.DataFrame(dashen_json)
    with open(r'..\data\scrape_results\boa\review_boa_bank_scrape_result.json', 'r', encoding='utf-8') as boa_data:
        boa_json = json.load(boa_data)
        df_boa = pd.DataFrame(boa_json)
    with open(r'..\data\scrape_results\cbe\review_cbe_bank_scrape_result.json', 'r', encoding='utf-8') as cbe_data:
        cbe_json = json.load(cbe_data)
        df_cbe = pd.DataFrame(cbe_json)
    return df_dashen, df_boa, df_cbe

df_dashen, df_boa, df_cbe = load_data()
    
```
